# 7. [10 pts] Attempt a few classifier models and report their 10-fold CV performance.


```python
X = df[[' Bwd Packet Length Mean', ' Avg Bwd Segment Size',
       'Bwd Packet Length Max', ' Bwd Packet Length Std',
       ' Packet Length Mean', ' Average Packet Size',
       ' Packet Length Std', ' Max Packet Length',
       ' Packet Length Variance']]
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
y = df[' Label']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn import metrics
from sklearn.model_selection import StratifiedKFold
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0,n_jobs=-1)
kfold = StratifiedKFold(n_splits=10).split(X, y)
scores = []
for k, (train, test) in enumerate(kfold):
     clf.fit(X[train], y[train])
     score = clf.score(X[test], y[test])
     scores.append(score)
     print('Fold: %2d, Acc: %.3f' % (k+1, score))
print("Mean Accuracy", np.mean(scores))
```

## Linear SVM

```python
from sklearn.svm import LinearSVC
clf = LinearSVC(random_state=0)
clf = RandomForestClassifier(random_state=0,n_jobs=-1)
kfold = StratifiedKFold(n_splits=10).split(X, y)
scores = []
for k, (train, test) in enumerate(kfold):
     clf.fit(X[train], y[train])
     score = clf.score(X[test], y[test])
     scores.append(score)
     print('Fold: %2d, Acc: %.3f' % (k+1, score))
print("Mean Accuracy", np.mean(scores))
```
